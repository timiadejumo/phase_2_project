## Graph is Hard to Read So We Remove Outliers

```python
# Define function to determine outliers
def outliers(df, col):
    Q1= df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    upper_limit = Q3 + 3 * IQR
    lower_limit = Q1 - 3 * IQR
    return upper_limit, lower_limit

# Apply outlier function to 'price' data
upper, lower = outliers(df, "price")
print("Upper whisker: ", upper)
print("Lower Whisker: ", lower)
```

```python
# Filter outliers with values from outlier function 
no_outlier_df = df[(df['price'] > lower) & (df['price'] < upper)]

# Create Seattle Dataframe for graphing purposes
SEA_no_outlier_df = no_outlier_df[(no_outlier_df['Region'] == 'SEA_N') | 
                                 (no_outlier_df['Region'] == 'SEA_S') |
                                 (no_outlier_df['Region'] == 'SEA_W') |
                                 (no_outlier_df['Region'] == 'SEA_E')]
# Create Non-Seattle Dataframe for graphing purposes
nonSEA_no_outlier_df = no_outlier_df[(no_outlier_df['Region'] == 'North') | 
                                 (no_outlier_df['Region'] == 'South') |
                                 (no_outlier_df['Region'] == 'East') |
                                 (no_outlier_df['Region'] == 'West')]

# Verify no data was lost
len(SEA_no_outlier_df) + len(nonSEA_no_outlier_df) == len(no_outlier_df)
```
