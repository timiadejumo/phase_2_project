# Chris' Code

```python
# Convert date col to datetime with pandas
df['quarter'] = pd.to_datetime(df['date'])

# Convert to quarters
df['quarter'] = pd.PeriodIndex(df.date, freq='Q')
```

```python
# Check value counts
df['quarter'].value_counts()
```

```python
# Cast as str type for 
df['quarter'] = df['quarter'].astype(str)
```

```python
# Converting str to numeric to for ordering
df['quarter'] = df['quarter'].str.replace('2014Q2', '1').replace('2014Q3', '2').replace('2014Q4', '3').replace('2015Q1', '4').replace('2015Q2', '5')

# Convert to int
df['quarter'] = df['quarter'].astype(int)
```

# Dummy Regressor (in progress)

```python
y = df.price
X = df[['zipcode', 'quarter', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.20, random_state = 42)
```

```python
dummy_reg = DummyRegressor(strategy="mean")
dummy_reg.fit(X_train, y_train)
dummy_reg.predict(X_train)
dummy_reg.score(X_train, y_train)
```
